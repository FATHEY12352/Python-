**1 Алфавит**

Напишите программу, выводящую следующий список:
['a', 'bb', 'ccc', 'dddd', 'eeeee', 'ffffff', ...]
Последний элемент списка должен состоять из 26 символов z.

```python
# سلسلة الأحرف المعطاة
abc = 'abcdefghijklmnopqrstuvwxyz'

# 1. نستخدم "تعبير القائمة" (List Comprehension)
# i ستمثل الموقع (0 للحرف a, 1 للحرف b, إلخ)
# abc[i] هو الحرف في الموقع i
# (i + 1) هو عدد المرات التي نريد تكرار الحرف (1, 2, 3...)
alphabet_list = [abc[i] * (i + 1) for i in range(len(abc))]

# 2. نطبع القائمة
print(alphabet_list)
```

**2 Без дубликатов**

На вход программе подается натуральное число n, а затем n строк. Напишите программу, которая выводит только уникальные строки, в том же порядке, в котором они были введены.

```python
try:
    n = int(input("أدخل عدد الأسطر n: "))

    # قائمة لتخزين الأسطر الفريدة التي ظهرت
    unique_lines = []

    for _ in range(n):
        line = input()

        # 1. نتحقق إذا كان السطر "ليس موجوداً" في قائمتنا
        if line not in unique_lines:
            # 2. إذا كان جديداً، نضيفه إلى القائمة
            unique_lines.append(line)

    # 3. نطبع كل عنصر فريد في القائمة
    for line in unique_lines:
        print(line)

except ValueError:
    print("خطأ: يرجى إدخال عدد صحيح لـ n.")
```

**3 Отрицательные, нули и положительные**

На вход программе подается натуральное число n, а затем n целых чисел. Напишите программу, которая сначала выводит все отрицательные числа, затем нули, а затем все положительные числа, каждое на отдельной строке. Числа должны быть выведены в том же порядке, в котором они были введены.

```python
try:
    n = int(input("أدخل عدد الأرقام n: "))

    # 3 قوائم لتصنيف الأرقام
    negatives = []
    zeros = []
    positives = []

    # 1. نقرأ الأرقام ونصنفها
    for _ in range(n):
        num = int(input())
        if num < 0:
            negatives.append(num)
        elif num == 0:
            zeros.append(num)
        else:
            positives.append(num)

    # 2. نطبع القوائم بالترتيب المطلوب
    # (كل رقم في سطر منفصل)
    for num in negatives:
        print(num)

    for num in zeros:
        print(num)

    for num in positives:
        print(num)

except ValueError:
    print("خطأ: يرجى إدخال أعداد صحيحة.")
```

**4 Более одного экземпляра**

На вход программе подается строка, содержащая целые неотрицательные числа, разделенные пробелом. Напишите программу, которая выводит те числа, которые встречаются в данной последовательности более одного раза.

```python
# 1. نقرأ السطر ونقسمه إلى قائمة من "النصوص"
numbers_str = input().split()

duplicates = [] # لتخزين المكررات (مرة واحدة)

# 2. نمر على كل رقم "نصي" في القائمة
for num in numbers_str:
    # 3. نتحقق من شرطين:
    #    أ) هل تكرر الرقم أكثر من مرة؟ (numbers_str.count(num) > 1)
    #    ب) هل أضفناه "بالفعل" إلى قائمة المكررات؟ (num not in duplicates)
    if numbers_str.count(num) > 1 and num not in duplicates:
        duplicates.append(num)

# 4. نطبع المكررات التي وجدناها
print(' '.join(duplicates))
```

**5 Суммы соседних чисел**

На вход программе подается натуральное число n, где n≥2. Затем поступают
n целых чисел. Напишите программу, которая создает из указанных чисел список, состоящий из сумм соседних чисел (первого и второго, второго и третьего и т.д.)

```python
try:
    n = int(input("أدخل عدد الأرقام n (n >= 2): "))

    numbers = []
    # 1. نقرأ كل الأرقام ونضعها في قائمة
    for _ in range(n):
        numbers.append(int(input()))

    # 2. نستخدم تعبير القائمة لحساب المجاميع
    # (نحن نتوقف عند العنصر "قبل الأخير"، لذا نستخدم range(n - 1))
    adjacent_sums = [numbers[i] + numbers[i+1] for i in range(n - 1)]

    # 3. نطبع القائمة الناتجة
    print(adjacent_sums)

except (ValueError, IndexError):
    print("خطأ: تأكد من إدخال n >= 2 وأعداد صحيحة.")
```

**6 Переставить минимум и максимум**

На вход программе подается строка текста, содержащая **различные** натуральные числа. Вам необходимо переставить максимальный и минимальный элементы местами и вывести измененную **строку**.

```python
try:
    # 1. نقرأ السطر، نقسمه، ونحوله إلى قائمة أعداد صحيحة
    line = input()
    nums = [int(n) for n in line.split()]

    # 2. نجد القيمة الصغرى والكبرى
    min_val = min(nums)
    max_val = max(nums)

    # 3. نجد "موقع" (index) كل منهما
    min_index = nums.index(min_val)
    max_index = nums.index(max_val)

    # 4. نقوم بالتبديل السحري في بايثون!
    nums[min_index], nums[max_index] = nums[max_index], nums[min_index]

    # 5. نحول الأرقام مرة أخرى إلى نصوص لطباعتها
    output_str = [str(n) for n in nums]
    print(' '.join(output_str))

except (ValueError):
    print("خطأ في الإدخال.")
```

**7 Количество артиклей**

На вход программе подается строка, содержащая английский текст. Слова текста разделены символом пробела. Напишите программу, которая подсчитывает общее количество артиклей: 'a', 'an', 'the'. Артикли могут начинаться с заглавной буквы.

```python
# النص المعطى كمثال
string = 'William Shakespeare was born in the town of Stratford, England, in the year 1564. When he was a young man, Shakespeare moved to the city of London, where he began writing plays. His plays were soon very successful, and were enjoyed both by the common people of London and also by the rich and famous. In addition to his plays, Shakespeare wrote many short poems and a few longer poems. Like his plays, these poems are still famous today.'

# يمكنك إلغاء التعليق عن السطر التالي لقراءة نص من المستخدم
# string = input("أدخل النص: ")

# 1. نحول النص كله إلى أحرف صغيرة ونقسمه إلى قائمة كلمات
words = string.lower().split()

# 2. نعد كل أداة تعريف
count_a = words.count('a')
count_an = words.count('an')
count_the = words.count('the')

# 3. نجمع العدد الإجمالي
total_articles = count_a + count_an + count_the

print(f"إجمالي عدد الأدوات (a, an, the): {total_articles}")
```

**8 Группировка дубликатов**

На вход программе подается строка текста, содержащая символы, разделенные пробелом. Напишите программу, которая упаковывает последовательности одинаковых символов заданной строки в подсписки.

```python
# 1. نقرأ ونقسم الرموز
chars = input().split()

# 2. قائمة النتائج النهائية
result = []

# 3. نتأكد أن القائمة ليست فارغة
if len(chars) > 0:

    # 4. نبدأ أول مجموعة بالعنصر الأول
    current_group = [chars[0]]

    # 5. نمر على "باقي" العناصر (من العنصر الثاني)
    for i in range(1, len(chars)):

        # 6. نقارن العنصر الحالي (i) بالسابق (i-1)
        if chars[i] == chars[i-1]:
            # إذا كانا متطابقين، أضفه للمجموعة الحالية
            current_group.append(chars[i])
        else:
            # إذا كانا مختلفين:
            # 1. أضف المجموعة "القديمة" المكتملة إلى النتائج
            result.append(current_group)
            # 2. ابدأ مجموعة "جديدة" بالعنصر الحالي
            current_group = [chars[i]]

    # 7. لا تنسَ إضافة "آخر" مجموعة بعد انتهاء الحلقة
    result.append(current_group)

# 8. طباعة النتيجة
print(result)
```

**9 Одной строкой_1**

На вход программе подается строка текста. Напишите программу, использующую **списочное выражение**, которая выводит все цифровые символы данной строки.

```python
text = input("أدخل النص: ")

# 1. نستخدم تعبير القائمة لفلترة الأرقام فقط
digits = [char for char in text if char.isdigit()]

# 2. نجمع الأرقام في نص واحد ونطبعها
# (مثلاً لو كانت القائمة ['1', '2', '3']، .join يحولها إلى "123")
print(''.join(digits))
```

**10 Одной строкой_2**

На вход программе подаётся строка текста, содержащая целые числа. Напишите программу, использующую **списочное выражение**, которая выведет квадраты чётных чисел, кроме тех квадратов, которые оканчиваются на цифру 4.

```python
try:
    # 1. نقرأ السطر، نقسمه، ونحوله إلى قائمة أعداد صحيحة
    nums = [int(n) for n in input().split()]

    # 2. تعبير القائمة
    result = [
        n * n                  # 1. ماذا نريد: مربع العدد
        for n in nums          # 2. المصدر: لكل عدد n في قائمتنا
        if n % 2 == 0          # 3. الفلتر الأول: إذا كان العدد زوجياً
        if (n * n) % 10 != 4   # 4. الفلتر الثاني: وإذا كان (مربعه % 10) لا يساوي 4
    ]

    # 3. نطبع النتيجة (الأرقام مفصولة بمسافة)
    print(*result)

except ValueError:
    print("خطأ: يرجى إدخال أعداد صحيحة مفصولة بمسافة.")
```

**11 Количество слов в тексте**

Напишите программу для определения общего количества **различных** слов в строке текста без учета регистра.

Примечания:

* Словом считается последовательность непробельных символов, идущих подряд, слова разделены одним или большим числом пробелов.
* Знаками препинания '.,;:-?!' пренебрегаем, других знаков в тексте не будет.



```python
text = input("أدخل النص: ")
punctuation = '.,;:-?!'

# 1. نحول النص إلى أحرف صغيرة
text_lower = text.lower()

# 2. نزيل كل علامات الترقيم
for p in punctuation:
    text_lower = text_lower.replace(p, '') # استبدلها بـ "لا شيء"

# 3. نقسم النص النظيف إلى قائمة كلمات
words = text_lower.split()

# 4. نحول قائمة الكلمات إلى "مجموعة" (Set)
# هذا سيزيل كل التكرار تلقائياً
unique_words = set(words)

# 5. نطبع "حجم" المجموعة (أي عدد العناصر الفريدة)
print(f"عدد الكلمات الفريدة: {len(unique_words)}")
```

**12 Оригиналы и дубликаты**

На вход программе подается строка текста, содержащая числа. Для каждого числа выведите слово YES, если это число ранее встречалось в последовательности или NO, если не встречалось.

**Примечание.** Ведущие нули в числах должны игнорироваться.

```python
# 1. مجموعة لتخزين الأعداد التي "رأيناها"
seen = set()

# 2. قائمة لتخزين النتائج (YES/NO)
result = []

# 3. نقرأ السطر ونقسمه
numbers_str = input().split()

# 4. نمر على كل "نص رقمي"
for num_str in numbers_str:

    # 5. نحول النص إلى رقم (int)
    # هذا يعالج مشكلة الأصفار البادئة (int('002') == 2)
    num = int(num_str)

    # 6. نتحقق: هل الرقم موجود "بالفعل" في مجموعة "seen"؟
    if num in seen:
        # نعم، لقد رأيناه من قبل
        result.append('YES')
    else:
        # لا، هذه أول مرة نراه
        result.append('NO')
        # والآن نضيفه إلى "seen" للمرات القادمة
        seen.add(num)

# 7. نطبع كل النتائج مفصولة بمسافة
print(' '.join(result))
```
